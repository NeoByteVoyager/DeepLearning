# 机器学习(Linear model)

## 一.什么是机器学习

- 机器学习=一个函数

## 二.不同函数的类型(机器学习的任务)


- 回归(regression)：函数输出一个标量(scalar)
- 分类(classification)：给定选项，机器输出正确的选项
- structure learning:create something with structure(image、ducument)

## 三.寻找函数的三个步骤

1. Function with unknown parameters(model)
2. Define loss from training data
3. Optimization

> 函数的选择取决于对原始问题和数据的分析

> loss：is a function of parameters，evaluate how good a set of values is   
> types：MAE(mean absolute error):$e=|y-\hat{y}|$,MSE(mean square error):$e=(y-\hat{y})^2$

# 深度学习

## 一.为什么需要深度模型

image.png

如果只有线性模型，那么我们只能得到蓝色的线性模型，如果真实数据是像红色线一样的数据，线性模型拟合不出来，没有办法模拟真实状况


## 二.怎么实现深度模型

一个真实场景下的数据可能是红色曲线分布

image.png

红色线都可以看作有限个蓝色线加上常量

image.png

对于上面的非线性情况，可以将两点间的距离取小点，看作线性情况

image.png

可以采用sigmod函数来模拟出不同的蓝色线

image.png

修改sigmod的参数，即可得到不同形状的蓝色线

image.png

## 三.深度模型的函数形式

**原始模型:** $y=b+wx_1$ → **深度模型：**$y=b+\sum_{i}ci*sigmod(b_i+w_ix_1)$   
**原始模型：**$y=b+\sum_{j}w_jx_j$ → **深度模型:**$y=b+\sum_{i}ci*sigmod(b_i+\sum_jw_{ij}x_j)$

image.png

- **feature：** x
- **unknown parameters：** $b,c^ {T},b,W$
- **θ：** all unknown parameters


## 四.深度模型的训练


- 深度模型的`Loss`和`Linear model`的的损失函数是一样的
- `optimization`的过程也是一样的
- 专业术语：  
  - initial values： $\boldsymbolθ^0$
  - gradient：$\boldsymbol g=∇L(\boldsymbol{θ}^0)$
  - update: $\boldsymbol{θ}^1 ←\boldsymbol{θ}^0 - α\boldsymbol g$
