## 一.目的

根据输入输出数据,寻找一个函数(function),**$f(输入)=一个标量(sclar)$**,且标量等于输出

## 二.步骤

1. 寻找`model`(a set of functions)
2. 定义评价函数的好坏的函数，也就是定义损失函数(eg:$L(f)=L(w,b)$)
3. 寻找最优函数:
  $$f^{*}=\arg\min_{f}L(f)$$
  或:$$w^*,b^*=\arg\min_{w,b}L(w,b)$$

## 三.模型选择

1. 规律比较简单时，可以直接使用使用简单线性模型:$y=wx+b$
2. 规律较为复杂时，可以使用二次、三次……：$y=w_1x+w_2x^2+b$……
3. 如果使用较为复杂的四次、五次，可能会让模型**过拟合(overfitting)**，在测试集上表现较差


## 四.优化模型选择

**重构模型：** 引入`domain knowledge`,在预测**不同类别的事物**时，不能一味使用相同的函数(模型)，可以对不同类别使用不同的模型：
$$\begin{aligned}
y = & \, b_1 \cdot \delta(x_s = \text{Pidgey}) \\
& + w_1 \cdot \delta(x_s = \text{Pidgey})x_{cp} \\
& + b_2 \cdot \delta(x_s = \text{Weedle}) \\
& + w_2 \cdot \delta(x_s = \text{Weedle})x_{cp} \\
& + b_3 \cdot \delta(x_s = \text{Caterpie}) \\
& + w_3 \cdot \delta(x_s = \text{Caterpie})x_{cp} \\
& + b_4 \cdot \delta(x_s = \text{Eevee}) \\
& + w_4 \cdot \delta(x_s = \text{Eevee})x_{cp}
\end{aligned}$$
以 $\text{Pidgey}$ 为例，指示函数 $\delta$ 的定义如下：
$$\delta(x_s = \text{Pidgey}) =
\begin{cases}
1 & \text{If } x_s = \text{Pidgey} \\
0 & \text{otherwise}
\end{cases}$$
由于`domain knowledge`限制，我们不知道**那些特征**会影响最后结果，可以把所有特征都加到模型中：  

If $x_s = \text{Pidgey}$:$$y' = b_1 + w_1 \cdot x_{cp} + w_5 \cdot (x_{cp})^2$$If $x_s = \text{Weedle}$:$$y' = b_2 + w_2 \cdot x_{cp} + w_6 \cdot (x_{cp})^2$$If $x_s = \text{Caterpie}$:$$y' = b_3 + w_4 \cdot x_{cp} + w_7 \cdot (x_{cp})^2$$If $x_s = \text{Eevee}$:$$y' = b_4 + w_4 \cdot x_{cp} + w_8 \cdot (x_{cp})^2$$
得到最终的完整预测公式：$$\begin{aligned}
y = & \, y' + w_9 \cdot x_{hp} + w_{10} \cdot (x_{hp})^2 \\
& + w_{11} \cdot x_h + w_{12} \cdot (x_h)^2 \\
& + w_{13} \cdot x_w + w_{14} \cdot (x_w)^2
\end{aligned}$$
但是这又会使模型变得异常复杂，又会导致**过拟合**，再测试数据上表现极差

## 五.优化损失函数（正则化Regularization）


1. 在损失函数后面加一个权值平方和项，要求权值越小越好: $$L=\sum _n(\hat y^n-(b+\sum w_ix_i ))^2+\lambda \sum (w_i)^2$$
- 权值越小，**函数就会越平滑**(输入改变时，输出改变较小,输出对输出不敏感)
- 数据有噪声的话，噪声对模型的影响较小
2. 对正则化的理解：
  - 不把偏置`b`加入：
    - 只有$w_i$决定曲线的斜率(平滑度),偏执$b$只影响曲线**向上向下**移动
  - 对$\lambda$的理解:
    - $\lambda$越大：函数越平滑，如果太大，整个权值平方和占比比较大，模型可能让所有的权值都趋于0,导致模型**欠拟合**
    - $\lambda$越小：如果它趋于0，和没使用这一项一样，会导致模型**过拟合**
    - 训练模型时，要不到调整$\lambda$,找到哪个既不过拟合又不欠拟合的点
