### 循环控制

```python
# for循环
for i in range(5):
  print(i)
# 打印数组元素
language=["C++","Verilog","Python"]
for lang in language:
  print(f"正在学习{lang}")
# 如果要用到索引
for i,lang in enumerate(language):
  print(f"正在学习{lang}，索引是{i}")
# while
count=0
while(count<3):
  print(count)
  count+=1
```

### 函数

```python
# 函数定义
def add(a,b):
  return a+b
print(add(10,4))
# 函数的参数和返回值是不固定的
def multiply_by_two(a):
  return a*2
print(multiply_by_two(3.14))
print(multiply_by_two("good"))
print(multiply_by_two([1,2]))

# 函数进阶，类型指定
def add_num(a:int,b:int)->dict:
  return {"name":"mr_wang"};
d=add_num(1,2)
print(d["name"])

```

## 面向对象基础



```python
# 类的创建
class Student:
  # 相当于C++的构造函数
  def __init__(self,name:str,score:int):
    # 必须挂上self
    self.name=name
    self.score=score
  def study(self,hours:int):
    print(f"{self.name}学习了{hours}小时，当前成绩{self.score}")
s1=Student("mr_wang",8)
s1.study(4)

# 继承与super（）
# 假设这是一个基础类模型
class Module:
  def __init__(self):
    self.is_training=True
class Myneural(Module):
  def __init__(self,input_size,hidden_size):
    # 必须先调用父类的构造函数
    super().__init__()

    # 定义自己的网络层
    self.input_size=input_size
    self.hidden_size=hidden_size
    print(f"底层训练状态，{self.is_training}")
m=Myneural(23,45)

# 重载
class addOperate:
  def __init__(self,bias):
    self.bias=bias
  def __call__(self,x):
    print("正在进行黑盒操作...")
    return self.bias+x

op=addOperate(10)

res=op(100)
print(res)
```
