### 变量与输出


```python
print("helloworld")
# 定义变量
user_name="mr_wang"
age=20
is_active="True"
# f_string 格式化输出
print(f"你好我是{user_name},今年{age}岁。")
```

### 条件判断


```python
# 输入（输入的结果都是字符串）
score = int(input("请输入你的分数:"))

if score>90:
  print("A")
elif score>80:
  print("B")
else:
  print("C")
```

### 循环控制

```python
# for循环
for i in range(5):
  print(i)
# 打印数组元素
language=["C++","Verilog","Python"]
for lang in language:
  print(f"正在学习{lang}")
# 如果要用到索引
for i,lang in enumerate(language):
  print(f"正在学习{lang}，索引是{i}")
# while
count=0
while(count<3):
  print(count)
  count+=1
```

### 常用数据筐：字典和列表


```python
# 列表
my_list=[10,20,30]
my_list.append(40)
print(len(my_list))
print(my_list[0])

# 字典
my_info={
    "name":"mr_wang",
    "school":"HNU",
    "score":"90"
}
print(my_info["name"])
my_info["school"]="CSU"
print(my_info["school"])
# 字典的键值对的数据类型并不是都是一样的
my_info["age"]=20
print(my_info["age"])
my_info[90]=80
print(my_info[90])

```

### 函数

```python
# 函数定义
def add(a,b):
  return a+b
print(add(10,4))
# 函数的参数和返回值是不固定的
def multiply_by_two(a):
  return a*2
print(multiply_by_two(3.14))
print(multiply_by_two("good"))
print(multiply_by_two([1,2]))

```
