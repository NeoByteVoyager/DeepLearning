## 2.LoRA

image.png

image.png

- **Insight:** Fine-tuning is changing the weights   
- **hypothesis:** weights matrix is low rank

变化矩阵：
$$
\Delta W=BA
$$

其中：

$$
A\in \mathbb{R}^{r\times d}
$$

$$
B\in \mathbb{R}^{d\times r}
$$

并且：

$$
r\ll d
$$
