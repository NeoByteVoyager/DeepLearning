# 二.Autograd

## Switch


**switch:**`requires_grad=True`  
It send a message to the autograd engine "This is a parameter.From now on,**track every single operation** that happens to it."

## Data vs Parameter

```python
import torch
x_data = torch.tensor([[1.,2.],[3.,4.]])
w = torch.tensor([[1.],[2.]],requires_grad=True)
print(f"Data tensor requires_grad:{x_data.requires_grad}")
print(f"Parameter tensor requires_grad:{w.requires_grad}")
```

## Computation Graph & .grad_fn

```python
a=torch.tensor(2.0,requires_grad=True)
b=torch.tensor(3.0,requires_grad=True)
c=torch.tensor(4.0,requires_grad=True)

y=a*b

z=y+c

print(z)
print(f"grad_fn for z:{z.grad_fn}")
print(f"grad_fn for y:{y.grad_fn}")
print(f"grad_fn for a:{a.grad_fn}")
```

## Dynamic selection


### 1.argmax

```python
x = torch.tensor([1.2, 5.5, 3.1, 0.8])
idx=x.argmax()
print(idx)
x = torch.tensor([[1.0, 6.0, 3.0],
          [4.0, 2.0, 5.0]])
print(x.argmax(dim=0))
print(x.argmax(dim=1))

# Practice:possibility to class
# 行代表：第几张图片； 列代表：它是[猫, 狗, 鸟]的概率分数
outputs = torch.tensor([[0.1, 0.8, 0.1],   # 预测第1张图
              [0.7, 0.2, 0.1],   # 预测第2张图
              [0.3, 0.3, 0.4]])  # 预测第3张图

# 真实标签（Ground Truth）
labels = torch.tensor([1, 0, 2]) # 分别是：狗、猫、鸟
# prediction
preds=outputs.argmax(dim=1)
print(preds)
acc= (preds==labels).float().mean()
print(f"accuracy:{acc}")
```

# 五.Build a simple model

```python
import torch
# 训练数据
X=torch.tensor([[1.2240], [-1.8022], [-1.3247], [-0.2105], [-0.3388]])
Y=torch.tensor([[3.4480], [-2.6044], [-1.6495],  [0.5790],  [0.3225]])
# 参数
W=torch.randn(1,1,requires_grad=True)
b=torch.randn(1,requires_grad=True)


# 设置超参数
alpha,epoches=0.01,1000

for epoch in range(epoches):
  Y_hat=X@W+b
  # 前向传播
  loss=((Y-Y_hat)**2).mean()

  # 反向传播
  loss.backward()
  with torch.no_grad():
    W-=alpha*W.grad
    b-=alpha*b.grad
    if epoch%10==0:
      print(f"epoch:{epoch:4d} ： loss:{loss.item():.4f} | W:{W.item():.4f} | b:{b.item():.4f}")
    W.grad.zero_()
    b.grad.zero_()
```

## 六.Torch.nn

- `torch.nn.Module`是一个基类("模板"),是**一个框架**,可以规定如何初始化`__init__`(网络的形状)以及`forward`(网络是如何流动的)
- `torch.nn.Linear`(全连接层)只是一个零件,自动创建`parameter`，并初始化

```python
import torch
import torch.nn as nn
import torch.optim as optim
# 训练数据
X=torch.tensor([[1.2240], [-1.8022], [-1.3247], [-0.2105], [-0.3388]])
Y=torch.tensor([[3.4480], [-2.6044], [-1.6495],  [0.5790],  [0.3225]])
# 创建自己的模型
class MyModel(nn.Module):
  def __init__(self,in_features,out_features):
    super().__init__()
    self.linear=nn.Linear(in_features,out_features)
  def forward(self,x):
    return self.linear(x)
# 实例化模型
model=MyModel(1,1)
# 第一损失函数和优化器
criterion=nn.MSELoss()  # 定义损失函数，代替手写((Y-Y_hat)**2).mean()
optimizer=optim.SGD(model.parameters(),lr=0.01) # 参数交给优化器管理,并设置学习率
for epoch in range(1000):
  # 前向传播
  Y_hat=model(X)
  # 计算损失函数
  loss=criterion(Y_hat,Y)

  # 清空梯度
  optimizer.zero_grad()
  # 反向传播
  loss.backward()
  # 更新参数
  optimizer.step()
'''
  if epoch%10==0:
    print(f"epoch:{epoch:4d}|loss:{loss.item():4f}|weights:{model.linear.weight.item():.4f}|bias:{model.linear.bias.item():.4f}")
'''
print(model)
```
